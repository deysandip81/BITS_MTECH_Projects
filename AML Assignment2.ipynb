```
import numpy as np
import matplotlib.pyplot as plt
```

## Visualization

```
# This function will help in visualization of our dataset. 
def plot_dataset(X, y, axes):
    plt.figure(figsize=(10,6))
    plt.plot(X[:, 0][y==0], X[:, 1][y==0], "bs",alpha = 0.5)
    plt.plot(X[:, 0][y==1], X[:, 1][y==1], "g^",alpha = 0.2)
    plt.axis(axes)
    plt.grid(True, which='both')
    plt.xlabel(r"$x_1$", fontsize=20)
    plt.ylabel(r"$x_2$", fontsize=20, rotation=0)
```

Our dataset contains two features $x_1$  and $x_2$, it has ten thousand instances with a Guassian noise of 0.4 and state = 42. it's a binary classification the labels are either 0 or 1.


```
from sklearn.datasets import make_moons

X, y = make_moons(n_samples=10000, noise=0.4, random_state=42)
```

##Visualize

```
plot_dataset(X, y, [-3, 5, -3, 3])
```

##Splitting


```
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2)
```

## Selecting the model with default parameters




```
from sklearn.tree import DecisionTreeClassifier

tree_clf = DecisionTreeClassifier()
```

**`Selecting`** the parameters for our GridSearch:


```
from sklearn.model_selection import GridSearchCV

parameter = {
             'criterion' : ["gini", "entropy"],

            }
```

Using a cross validation of 5 and measuring accuracy.

```
clf = GridSearchCV(tree_clf, parameter, cv = 5,scoring = "accuracy",return_train_score=True,n_jobs=-1)
```

Fitting the *model*

```
clf.fit(X_train, y_train)
```


 **Reporting & Analysis**

```
 
print("Run Report"),
print("-------------------"),
print("Train Accuracy: "),
print(clf.score(X_train, y_train))
print()
print("Test Accuracy:"),
print(clf.score(X_test, y_test)),
print()
```
